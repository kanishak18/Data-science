```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("dark_background")
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x = "State/UTs", y = "Total Cases", data = covid);
```
